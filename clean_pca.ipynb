```python
#载入包
import numpy as np
import pandas as pd
#用于写入excel
import xlwt
```

```python
#导入数据
raw_data = pd.read_excel('pcadata.xlsx')
raw_data.shape
#去除空值的行
data_dropna = raw_data.dropna(axis=0,how='any')
data_dropna.to_excel('d:\\该干正事儿\我在上海师范大学\争取顺利毕业\中期答辩\数据\pca_data_dropna.xlsx', encoding='utf-8', index=False, header=True)
data_dropna.shape
```

```python
from sklearn import preprocessing
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
data = data_dropna.iloc[:,7:24]
data
```

```python
#建立对象
minmax = preprocessing.MinMaxScaler()
#标准化处理
data_minmax = minmax.fit_transform(data)
pd.DataFrame(data_minmax).to_excel('d:\\该干正事儿\我在上海师范大学\争取顺利毕业\中期答辩\数据\pca_minmax.xlsx', encoding='utf-8', index=False, header=True)  #保存结果
```

```python
pca = PCA()   #保留所有成分
pca.fit(data_minmax)
pca.components_ #返回模型的各个特征向量
pca.explained_variance_ratio_ #返回各个成分各自的方差百分比(也称贡献率）
```

```python
pca = PCA(4)   #保留所有成分
pca.fit(data_minmax)
pca.components_ #返回模型的各个特征向量
```

```python
low_d = pca.transform(data)   #降低维度
pd.DataFrame(low_d).to_excel('d:\\该干正事儿\我在上海师范大学\争取顺利毕业\中期答辩\数据\pca_result.xlsx', encoding='utf-8', index=False, header=True)  #保存结果
```
